## 2. WEB-SCRAPPING REVIEWS AND RATINGS
Perform web-scrapping and create a DataFrame of amazon reviews on its latest smart phone OnePlus8. Here is your link to scrap the customer reviews, ratings.
https://www.amazon.in/OnePlus-Onyx-Black-128GB-Storage/product-reviews/B071Z97T2C/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews


```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd

path_chromedriver = r'C:\etc\chromedriver.exe'

url = 'https://www.amazon.in/OnePlus-Onyx-Black-128GB-Storage/product-reviews/B071Z97T2C/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'

driver = webdriver.Chrome(path_chromedriver)  # Optional argument, if not specified will search path.

####
# xpath data
####

xpath_usernames = "//div[@id='cm_cr-review_list']/div[@data-hook='review']//div[@class='a-profile-content']/span[@class='a-profile-name']"

xpath_titles =    "//div[@id='cm_cr-review_list']/div[@data-hook='review']//a[@data-hook='review-title']"

xpath_ratings =   "//div[@id='cm_cr-review_list']/div[@data-hook='review']//div[@class='a-row']//a[@class='a-link-normal']"

xpath_stylename = "//div[@id='cm_cr-review_list']/div[@data-hook='review']//div[@class='a-row a-spacing-mini review-data review-format-strip']//a[@class='a-size-mini a-link-normal a-color-secondary']"

xpath_date =      "//div[@id='cm_cr-review_list']/div[@data-hook='review']//span[@data-hook='review-date']"
```

```python
#cm_cr-review_list
driver = webdriver.Chrome(path_chromedriver)  # Optional argument, if not specified will search path.

driver.get(url)
df_lst=list()
pagesource_lst = list()
usernames_lst = list()
pageNum = 1

tst=0 # flag for testing/development to cancel page-wise iteration

while True:
    try:
        print(pageNum)
        # Get data
        usernames = driver.find_elements_by_xpath(xpath_usernames)
        
        username_seri = pd.Series([username.get_attribute('innerHTML') for username in usernames])
        
        titles = driver.find_elements_by_xpath(xpath_titles)
        title_seri = pd.Series([title.text for title in titles])
        
        ratings  = driver.find_elements_by_xpath(xpath_ratings)        
        ratings_seri = pd.Series([int(r.get_attribute('title')[0]) for r in ratings])
        
        stylename = driver.find_elements_by_xpath(xpath_stylename)
        stylename_seri = pd.Series([s.text[12:s.text.find(' StorageColour')] for s in stylename])
        colour_seri = pd.Series([s.text[s.text.find('StorageColour')+15:]for s in stylename])
        
        date_geo = driver.find_elements_by_xpath(xpath_date)      
        geo_seri = pd.Series([d_g.text[12:d_g.text.find(" on ")] for d_g in date_geo])
        date_seri = pd.Series([d_g.text[d_g.text.find(" on ")+4:] for d_g in date_geo])
        date_seri = pd.to_datetime(date_seri)
        # Create df & append
        df = pd.DataFrame()
        df['username'] = username_seri
        df['title'] = title_seri
        df['rating'] = ratings_seri
        df['stylename'] = stylename_seri
        df['colour'] = colour_seri
        df['geo'] = geo_seri
        df['date'] = date_seri
        
        df_lst.append(df)
        #############
        # test flag to skip iteration
        if tst:
            break
        #############
        # click next
        next_button = driver.find_element_by_xpath("//ul[@class='a-pagination']/li[2]")
        
        if 'a-disabled' in next_button.get_attribute('class'):
            print("finished paging")
            break
        next_button.click()

        #############
        pageNum+=1
        time.sleep(1)
        
    except Exception as e:
        print(f"{e}, {type(e)}")
        break
```

```python
big_df = pd.concat(df_lst)
big_df.reset_index(inplace=True)
```
