```python
import cv2
#import imutils
import matplotlib.pyplot as plt

```

```python
# cv2 cascade init from xml files
face_cascade  = cv2.CascadeClassifier("/home/hadoop/haarcascade/haarcascade_frontalface_default.xml")
eye_cascade   = cv2.CascadeClassifier("/home/hadoop/haarcascade/haarcascade_eye.xml")
smile_cascade = cv2.CascadeClassifier("/home/hadoop/haarcascade/haarcascade_smile.xml")

```
